```python
def getSeeds(seeds):
    isRange = False
    seed = None

    allSeeds = []
    oldSeed = 0
    for seed in seeds:
        if isRange:
            end = oldSeed + int(seed) - 1
            allSeeds.extend(list(range(oldSeed, end + 1)))
            isRange = False

        else:
            oldSeed = int(seed)
            isRange = True
    return allSeeds
```
