```python
def fiveKind(cards):
    if len(set(cards)) == 1:
        return True
    return False
    
def fourKind(cards):
    for i in cards:
        if cards.count(i) == 4:
            return True
    return False

def fullHouse(cards):
    if len(set(cards)) == 2:
        return True
    return False

def threeKind(cards):
    for i in cards:
        if cards.count(i) == 3:
            return True
    return False

def twoPair(cards):
    if len(set(cards)) == 3:
        return True
    return False
            
def onePair(cards):
    for i in cards:
        if cards.count(i) == 2:
            return True
    return False
```

```python
hands = []

with open('./calibration document [d7p2].txt', 'r') as f:
    for line in f:
        hands.append(line.split())

strengths = {
}
```

```python
i = 0
for hand in hands:
    cards, bids = hand  
    strength = -1

    if(fiveKind(cards)):
        strength = 6

    # KKKKW
    # KKKKJ
    # JJJJK

    elif(fourKind(cards)):
        if "J" in cards:
            strength = 6
            
        else:
            strength = 5

    # KKQQQ
    # JJQQQ
    # QQJJJ

    elif(fullHouse(cards)):
        
        if "J" in cards and (cards.count("J") == 2 or cards.count("J") == 3):
            strength = 6

        else:
            strength = 4

    # KQTTT
    # JQTTT
    # Q36JJ

    elif(threeKind(cards)):
        if "J" in cards and (cards.count("J") == 1 or cards.count("J") == 3):
            strength = 5


        else:
            strength = 3
    
    # KKQQJ
    # KKJJ9
    # KKQQ5

    elif(twoPair(cards)):
        if "J" in cards and (cards.count("J") == 1):
            strength = 4
        elif "J" in cards and (cards.count("J") == 2):
            strength = 5
        
        else:
            strength = 2

    # KKQJ9
    # JJQKA
    
    elif(onePair(cards)):
        if "J" in cards and (cards.count("J") == 1):
            strength = 3

        elif "J" in cards and (cards.count("J") == 2):
            strength = 3

        else:
            strength = 1
        
    # KQ4AT

    else:
        if "J" in cards:
            strength = 1
            
        else:
            strength = 0

    num = [cards, bids]

    if strengths.get(strength) == None:
        strengths[strength] = []
    
    strengths[strength].append(num)
```

```python
list = ['J','2','3','4','5','6','7','8','9','T','Q','K','A']

def sortCards(sublist):
    string = sublist[0]
    return [list.index(char) for char in string]
```

```python
rank = len(hands)
ans = 0

for i in range(6,-1,-1):
    if strengths.get(i) == None:
        continue

    strengths[i] = sorted(strengths[i], key=sortCards, reverse=True)
```

```python
rank = len(hands)
ans = 0


for i in range(6,-1,-1):
    if strengths.get(i) == None:
        continue

    for hand in strengths.get(i):
        cards, bids = hand
        ans += (rank * int(bids))
        rank -= 1

print(ans)
```
