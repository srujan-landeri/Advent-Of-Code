```python
with open('./calibration document [d5p1].txt', 'r') as f:
    lines = f.readlines()

    # reeding seeds
    seeds = lines[0].split(':')[1].split()

    # seed-to-soil map
    seed_to_soil = []
    
    ind = 1
    while "seed-to-soil map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        seed_to_soil.append(lines[ind])
        ind+=1

    # soil to fert map
    soil_to_fertilizer = []
    
    while "soil-to-fertilizer map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        soil_to_fertilizer.append(lines[ind])
        ind+=1

    # soil to fert map
    fertilizer_to_water = []
    
    while "fertilizer-to-water map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        fertilizer_to_water.append(lines[ind])
        ind+=1

    # water-to-light map
    water_to_light = []
    
    while "water-to-light map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        water_to_light.append(lines[ind])
        ind+=1

    # light-to-temperature map:
    light_to_temperature = []
    
    while "light-to-temperature map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        light_to_temperature.append(lines[ind])
        ind+=1

    # temperature-to-humidity map:
    temperature_to_humidity = []

    while "temperature-to-humidity map:" not in lines[ind]:
        ind+=1
    ind+=1
    while len(lines[ind].strip()) != 0:
        temperature_to_humidity.append(lines[ind])
        ind+=1

    # humidity-to-location map:

    humidity_to_location = []

    while "humidity-to-location map:" not in lines[ind]:
        ind+=1
    ind+=1
    while ind != len(lines):
        humidity_to_location.append(lines[ind])
        ind+=1
```

```python
seeds = list(map(int, seeds))
blocks = [
    seed_to_soil,
    soil_to_fertilizer,
    fertilizer_to_water,
    water_to_light,
    light_to_temperature,
    temperature_to_humidity,
    humidity_to_location
]

print(seeds)
for block in blocks:
    ranges = []
    for range in block:
        ranges.append(list(map(int, range.split())))

    new = []

    for x in seeds:
        for a, b, c in ranges:
            if b <= x < b + c:
                new.append(x - b + a)
                break
        else:
            new.append(x)
    seeds = new

print(min(seeds))
```
