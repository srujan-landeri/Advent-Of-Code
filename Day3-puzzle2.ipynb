```python
with open('./calibration document [d3p2].txt', 'r') as f:
    lines = f.readlines()
```

```python
sum = 0

for i in range(1,len(lines)-1):
    for j in range(1,len(lines[i])-1):
        total_num = 0
        if lines[i][j] == '*':
            total_num = 0

            # checking for digit left and right side of the *
            if lines[i][j-1].isdigit():
                total_num += 1
                
            if lines[i][j+1].isdigit():
                total_num += 1

            # the above one is a dot
            if lines[i - 1][j] == '.':

                if lines[i-1][j-1].isdigit():
                    total_num += 1

                if lines[i-1][j+1].isdigit():
                    total_num += 1
            
            if lines[i - 1][j].isdigit():
                total_num += 1
        
            # the below one is a dot
            if lines[i + 1][j] == '.':

                if lines[i+1][j-1].isdigit():
                    total_num += 1

                if lines[i+1][j+1].isdigit():
                    total_num += 1
            
            if lines[i + 1][j].isdigit():
                total_num += 1

            if total_num != 2:
                continue

            # if two part numbers found

            else:
                num1 = None 
                num2 = None 

                # checking for digit left and right side of the *
                if lines[i][j-1].isdigit():
                    
                    end = j - 1
                    start = end

                    while lines[i][start].isdigit():
                        start -= 1
                    
                    num1 = int(lines[i][start+1:end+1])

                # chefcking for digit left and right side of the *
                if lines[i][j+1].isdigit():

                    start = j + 1
                    end = start

                    while lines[i][end].isdigit():
                        end += 1
                    
                    if num1 == None:
                        num1 = int(lines[i][start:end])
                    else:
                        num2 = int(lines[i][start:end])
                
                # the above one is a dot
                if lines[i - 1][j] == '.':

                    # checking left up part
                    if lines[i - 1][j - 1].isdigit():

                        end = j - 1
                        start = end

                        while lines[i - 1][start].isdigit():
                            start -= 1

                        if num1 is None:
                            num1 = int(lines[i - 1][start + 1:end + 1])

                        else:
                            num2 = int(lines[i - 1][start + 1:end + 1])

                    # checking right up part
                    if lines[i - 1][j + 1].isdigit():

                        start = j + 1
                        end = start

                        while lines[i - 1][end].isdigit():
                            end += 1

                        if num1 is None:
                            num1 = int(lines[i - 1][start:end])

                        else:
                            num2 = int(lines[i - 1][start:end])

                # getting the above number
                if lines[i - 1][j].isdigit():

                    end = j
                    start = end

                    while lines[i - 1][start].isdigit():
                        start -= 1

                    while lines[i - 1][end].isdigit():
                        end += 1

                    if num1 is None:
                        num1 = int(lines[i - 1][start + 1:end])

                    else:
                        num2 = int(lines[i - 1][start + 1:end])

                # the below one is a dot

                if lines[i + 1][j] == '.':
                    
                    # down left
                    if lines[i + 1][j - 1].isdigit():

                        end = j - 1
                        start = end

                        while lines[i + 1][start].isdigit():
                            start -= 1

                        if num1 is None:
                            num1 = int(lines[i + 1][start + 1:end + 1])

                        else:
                            num2 = int(lines[i + 1][start + 1:end + 1])

                    # down right
                    if lines[i + 1][j + 1].isdigit():

                        start = j + 1
                        end = start

                        while lines[i + 1][end].isdigit():
                            end += 1

                        if num1 is None:
                            num1 = int(lines[i + 1][start:end])

                        else:
                            num2 = int(lines[i + 1][start:end])

                # down number
                if lines[i + 1][j].isdigit():

                    end = j
                    start = end

                    while lines[i + 1][start].isdigit():
                        start -= 1

                    while lines[i + 1][end].isdigit():
                        end += 1

                    if num1 is None:
                        num1 = int(lines[i + 1][start + 1:end])

                    else:
                        num2 = int(lines[i + 1][start + 1:end])

                sum += (num1 * num2)

print(sum)
```
